```python
#代码结构习题
#K为整型，下面程序中两个while循环的次数为（A）
k = j = 1000
while k > 1: 
    k = k//3
    print(k)
while j > 1:
    j = j/3
    print(j)

#A.6 7   B.6 6   C.5 7   D.7 8 
# 解：k 的值的变化为 333 111 37 12 4 1 所以第一个循环执行了6次；
#而第二个循环执行到第六次时，j = 1.37174211 所以还会执行一次，所以是7次
```

#简答题
解释python中的三元表达式。

解：
在python中的三元表达式不需要？而是使用if …… else ……
三元表达式的语法为：A if 判断 else B 
三元表达式的执行过程为：如果 判断的结果为真，那么A部分被执行，如果 判断结构为假，那么B部分被执行
例如： 
x = 100 if 3<4>5 else 90
print(x)
执行的结果为
90

```python
import csv 
import numpy as np
import matplotlib.pyplot as plt
with open('company-sales.csv', 'rt') as fin:
    cin = csv.reader(fin)
    villains = [row for row in cin] 

b = np.array(villains) # 将列表转化为numpy类型
b_x = b[1:,0]
b_y = b[1:,-1]
y = [int(i) for i in b_y] # 由于读取的数据都是字符串类型的，所以要转换类型

plt.title('company total profit of each months')
plt.xlabel('month')
plt.ylabel('total profit')
plt.plot(b_x, y, 'r--',
         b_x, y, 'ko', linewidth=3)
plt.legend(['month'], loc=4)
plt.savefig('xueli-11.png')
plt.show()
```

```python
import csv 
import numpy as np
import matplotlib.pyplot as plt
with open('company-sales.csv', 'rt') as fin:
    cin = csv.reader(fin)
    villains = [row for row in cin] 

b = np.array(villains) 
b_month = b[1:, 0] # 月份
b_facecream = [int(i) for i in b[1:, 1]]
b_facewash = [int(i) for i in b[1:, 2]]
b_toothpaste = [int(i) for i in b[1:, 3]]
b_bathingsoap = [int(i) for i in b[1:, 4]]
b_shampoo = [int(i) for i in b[1:, 5]]
b_moisturizer = [int(i) for i in b[1:, 6]]

plt.title('sales date')
plt.xlabel('month')
plt.ylabel('sales units')
plt.plot(b_month, b_facecream, 'r-')
plt.plot(b_month, b_facewash, 'b--')
plt.plot(b_month, b_toothpaste, 'k-')
plt.plot(b_month, b_bathingsoap, 'y:')
plt.plot(b_month, b_shampoo, 'c-')
plt.plot(b_month, b_moisturizer, 'g--')
plt.legend(['facecream','facewash','toothpaste','bathingsoap','shampoo','moisturizer'], loc=2)

plt.savefig('xueli-22.png')
plt.show()
```

```python
import csv 
import numpy as np
import matplotlib.pyplot as plt
with open('company-sales.csv', 'rt') as fin:
    cin = csv.reader(fin)
    villains = [row for row in cin] 

b = np.array(villains) 
b_month = b[1:, 0] # 月份
b_facecream = [int(i) for i in b[1:, 1]]
b_facewash = [int(i) for i in b[1:, 2]]

plt.title('sales date of fasecream & facewash')
plt.xlabel('month')
plt.ylabel('sales units')
plt.bar(b_month, b_facecream, width=0.4, align='edge') # 使用align参数将一个条形图向右偏移一些
plt.bar(b_month, b_facewash, width=0.4)

plt.legend(['facecream','facewash'], loc=2)

plt.savefig('xueli-33.png')
plt.show()
```
