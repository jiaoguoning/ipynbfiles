```python
#代码结构习题
#K为整型，下面程序中两个while循环的次数为（A）
k = j = 1000
while k > 1: 
    k = k//3
    print(k)
while j > 1:
    j = j/3
    print(j)

#A.6 7   B.6 6   C.5 7   D.7 8 
# 解：k 的值的变化为 333 111 37 12 4 1 所以第一个循环执行了6次；
#而第二个循环执行到第六次时，j = 1.37174211 所以还会执行一次，所以是7次
```

#简答题
解释python中的三元表达式。

解：
在python中的三元表达式不需要？而是使用if …… else ……
三元表达式的语法为：A if 判断 else B 
三元表达式的执行过程为：如果 判断的结果为真，那么A部分被执行，如果 判断结构为假，那么B部分被执行
例如： 
x = 100 if 3<4>5 else 90
print(x)
执行的结果为
90

```python
import csv 
import numpy as np
import matplotlib.pyplot as plt
with open('company-sales.csv', 'rt') as fin:
    cin = csv.reader(fin)
    villains = [row for row in cin] 

b = np.array(villains) # 将列表转化为numpy类型
b_x = b[1:,0]
b_y = b[1:,-1]
y = [int(i) for i in b_y] # 由于读取的数据都是字符串类型的，所以要转换类型

plt.title('company total profit of each months')
plt.xlabel('month')
plt.ylabel('total profit')
plt.plot(b_x, y, 'r--',
         b_x, y, 'ko', linewidth=3)
plt.legend(['month'], loc=4)
plt.savefig('xueli-11.png')
plt.show()
```

```python
import csv 
import numpy as np
import matplotlib.pyplot as plt

with open('appendix.csv', 'rt') as fin:
    cin = csv.reader(fin)
    villains = [row for row in cin] 

dataset = np.array(villains)

year = sorted(list(set([int(i[6:]) for i in dataset[1:, 2]])))
years = sorted(list(set([i[6:] for i in dataset[1:, 2]])))

mit_year = np.array([i for i in dataset[1:, (0,2)] if i[0]=='MITx'])
har_year = np.array([i for i in dataset[1:, (0,2)] if i[0]=='HarvardX'])

mit_year = list([int(i[6:]) for i in mit_year[0:, 1]])
har_year = list([int(i[6:]) for i in har_year[0:, 1]])

num_count_mit, num_count_har = [], []
for i in year:
    num_count_mit.append(mit_year.count(i))
for i in year:
    num_count_har.append(har_year.count(i))

plt.xlabel('years')
plt.ylabel('counts corses')
plt.plot(years, num_count_mit, 'r-')
plt.plot(years, num_count_har, 'g-')
plt.legend(['MITx','HarvardX'], loc=2)
plt.savefig('xueli11-2.png')
plt.show()
```

```python
import csv 
import numpy as np
import matplotlib.pyplot as plt

with open('appendix.csv', 'rt') as fin:
    cin = csv.reader(fin)
    villains = [row for row in cin] 

dataset = np.array(villains)
class_set = list(set(list(dataset[1:, 5]))) # 总共有几种主题
years = sorted(list(set([i[6:] for i in dataset[1:, 2]])))
sample = dataset[1:, (2, 5, 8)]

all_data = list()
for title in class_set:
    data = list()
    for year in years:
        nums = list()
        for sam in sample:
            if sam[0][-4:]==year and title==sam[1]:
                nums.append(int(sam[2]))
        data.append(round(sum(nums)/len(nums)) if len(nums)!=0 else 0)
    all_data.append(data)

plt.xlabel('years')
plt.ylabel('avg_number')
plt.plot(years, all_data[0], 'r-')
plt.plot(years, all_data[1], 'b--')
plt.plot(years, all_data[2], 'k-')
plt.plot(years, all_data[3], 'y:')
plt.legend(class_set, loc=1)

plt.savefig('xueli11-4.png')
plt.show()
```

```python
import csv 
import numpy as np
import matplotlib.pyplot as plt

with open('appendix.csv', 'rt') as fin:
    cin = csv.reader(fin)
    villains = [row for row in cin] 

dataset = np.array(villains)
schools = list(set(dataset[1:, 0]))
class_set = list(set(list(dataset[1:, 5]))) # 总共有几种主题

mit_data = np.array([i for i in dataset[1:, (0, 5, -3)] if i[0]=='MITx'])
har_data = np.array([i for i in dataset[1:, (0, 5, -3)] if i[0]=='HarvardX'])


mit_avg, har_avg = [], []
for title in class_set:
    num = list()
    for sam in mit_data:
        if sam[1]==title:
            num.append(float(sam[2]))
    mit_avg.append(round(sum(num)/len(num),2) if len(num)!=0 else 0)

for title in class_set:
    num = list()
    for sam in har_data:
        if sam[1]==title:
            num.append(float(sam[2]))
    har_avg.append(round(sum(num)/len(num), 2) if len(num)!=0 else 0)

plt.xlabel('subject')
plt.ylabel('avg_%male')
plt.bar(class_set, mit_avg, width=0.4, align='edge')
plt.bar(class_set, har_avg, width=0.4)
plt.legend(schools, loc=1)
plt.savefig('xueli11-5.png')
plt.show()
```
